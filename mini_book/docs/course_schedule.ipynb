## Course schedule

Here is the current week-by-week schedule ðŸ“… . We may adjust as we go along. To get started, we're going to create the calendar of weeks for the course programmatically rather than manually!


```python
## import modules
import pandas as pd
import re
import numpy as np


## tell python to display output and print multiple objects
from IPython.display import display, HTML
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

## create range b/t start and end date
## of course 
start_date = pd.to_datetime("2022-08-24")
fall_recess = pd.to_datetime("2022-11-23")
end_date = pd.to_datetime("2022-11-30")
st_alldates = pd.date_range(start_date, end_date)

## subset to days in that range equal to Wednesday
st_w = st_alldates[st_alldates.day_name().isin(['Wednesday'])]

## create data frame with that information
st_dates = [re.sub("2022\\-", "", str(day.date())) for day in st_w] 
course_sched = pd.DataFrame({'Date': ["Wednesday " + i for i in st_dates]})

## three types of week indices
cond = [course_sched['Date'] == "Wednesday 11-23",
       course_sched['Date'] == "Wednesday 11-30"]
choices = ["Fall Recess",
          course_sched.index]
course_sched['Week'] = ["Week " + i if i != "Fall Recess"
                        else i 
                        for i in list(np.select(cond, choices, 
                                default = course_sched.index + 1))]

## display the resulting date sequence
display(course_sched)

## next block of code creates the
## actual content; can click "show"
## to see the underlying code
```
